```python
from timeit import default_timer

def esPrimo(n):
    if n<1: #numeros negativos no son primos. 
        return False
    elif n==2: #unico número par que es primo.
        return True
    else:
        for i in range(2,n):
            if n%i==0:
                return False
        return True


def num_prim(n):
    lista=[]
    if esPrimo(n)==True: #si el grupo G es de orden primo, eso quiere decir que cada subgrupo representa un grupo generador ciclico de G.
        for i in range(0,n):
            lista.append(i)
        return(lista)
    elif esPrimo(n)==False: #si no es primo el orden de G, entonces eso quiere decir que existen ciertos generadores que son primos.
        for i in range(0,n):
            if esPrimo(i)==True:
                lista.append(i)
        return lista
    else:
        return False

def cantnumlist(n):
    return len(num_prim(n))

def descomponer(n):
    lista=[]
    for i in range(2,n+1): #sirve para saber cuantos y cuales factores promos tiene el número.
        while n%i == 0:
            lista.append(i)
            n=n/i #sirve para mirar cuantas veces se repite el divisor, hasta que su resto de cero.
    return lista

def totalgene(n):
    a=dict(zip(descomponer(n),map(lambda x: descomponer(n).count(x),descomponer(n)))) #se cuentatn cuantas veces se tiene cierto factor primo. Se usa el zip para unir el numero de la lista con su conteo que se hace apartir de la lamba y la función map.
    b=len(a.keys())
    listanum=list(a.keys())
    listacoef=list(a.values())
    print(a)
    contador=1
    i=0
    while (i<=b-1):
        contador*=(((listanum[i])**(listacoef[i]))-((listanum[i])**(listacoef[i]-1))) #Va haciendo la multiplicación entre los factores.
        i+=1
    return contador

def menu():
    numero=int(input("Número"))
    print("1.Fuerza bruta")
    print("2.Lema 2")
    print("3.Lista factores generadores")
    print("4.Tiempos de ejecución")
    a_saber=int(input("¿Qué quiere saber?"))
    if a_saber==1 and numero>0:
        print(num_prim(numero))
    elif a_saber==1 and esPrimo(numero)==False:
        print("Es un número negativo")
    elif a_saber==2:
        print(totalgene(numero))
    elif a_saber==3:
        print(descomponer(numero))
    elif a_saber==4:
        inicio=default_timer()
        num_prim(numero)
        fin=default_timer()
        print("Tiempo de ejecución fuerza fruta"+str(fin - inicio))
        inicio2=default_timer()
        totalgene(numero)
        fin2=default_timer()   
        print("Tiempo de ejecución fi de Euler"+str(fin2 - inicio2))  
    else:
        print("Número no encontrado.")

menu()





#print(totalgene(numero))
#print(descomponer(numero))
#print(num_prim(numero))
#print(esPrimo(10))
#print(cantnumlist(numero))



```
